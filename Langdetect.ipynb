```python
data = pd.read_csv('reviews_sampled.csv')
```

```python
# write all indexes to non enlish text file
f= open("non_eng.txt","w+")
for i in range(0,len(data)):
    text = data['text'][i]
    try:
        if((text != '') & (langdetect.detect(text) != 'en')):
                f.write(str(i) + "\n")
                print(i,text)
    except:
        f.write( str(i) + '\n')
        print(i,text)
```

```python
f.close()
```

```python
lines = open('non_eng.txt').read().splitlines()
dropList =list(map(int, lines))
```

```python
len(dropList)
```

```python
# drop all non english reviews and write to new csv file
df_new = data.drop(data.index[dropList])
```

```python
df_new.to_csv('final_english_reviews.csv')
```
